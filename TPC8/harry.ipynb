```python
import re
from gensim.models import Word2Vec
from gensim.utils import tokenize
```

```python
filename1 = "Harry_Potter_Camara_Secreta-br.txt"
filename2 = "Harry_Potter_e_A_Pedra_Filosofal.txt"

with open(filename1, 'r', encoding='utf-8') as file:
    harry1 = file.read()
    
with open(filename2, 'r', encoding='utf-8') as file:
    harry2 = file.read()
    
```

### Concatenar os dois textos

```python
with open("Harry_Geral.txt", 'w', encoding='utf-8') as file:
    file.write(harry1 + harry2)
file.close()
```

```python
with open("Harry_Geral.txt", 'r', encoding='utf-8') as file:
    harry = file.read()
```

### Tratamento do documento

```python
linhas = harry.split("\n")

tokens = []

for linha in linhas:
    linha = list(tokenize(linha, lower=True))
    tokens.append(linha)
```

```python
model = Word2Vec(tokens, vector_size=100, window=5, min_count=1, sg=1, epochs=5)    

print(model.wv.most_similar("harry"))
```

```python
def get_word(word):
    try:
        return model.wv[word]
    except KeyError:
        print('The word "'+word+'" does not appear in this model.')
```

```python
print(get_word("dumbledore"))
print(get_word("relíquias"))
print(get_word("belatrix"))
```

```python
def guardar_modelo(modelo):
    modelo.save("modelos/harry.model")
    
def carregar_modelo():
    modelo = Word2Vec.load("modelos/harry.model")
    return modelo


guardar_modelo(model)
harry_wv = carregar_modelo()
```

## Verificar Similariedades

#### Funções para verificar palavras similares

```python
def verificar_similaridades(word1,word2,modelo):
    if word1 in modelo.wv.key_to_index and word2 in modelo.wv.key_to_index:
        return("%s \t %s \t %.2f" % (word1,word2, modelo.wv.similarity(word1,word2)))
    else:
        print("Pelo menos uma das palavras não está presente no modelo.")



def mais_similares(word,modelo):
    if word in modelo.wv.key_to_index:
        palavras = modelo.wv.most_similar(word)
        print("\n"+word.upper())
        for palavra,numero in palavras:
            print("%s \t %.2f" % (palavra,numero))
    else:
        print("Essa palavra não está presente no modelo")
    
```

```python
print(verificar_similaridades('grifinória', 'harry',harry_wv))
print(verificar_similaridades('varinha', 'voldemort',harry_wv))
print(verificar_similaridades('nariz', 'voldemort',harry_wv))
print(verificar_similaridades('vassoura', 'harry',harry_wv))
print(verificar_similaridades('quadribol', 'bola',harry_wv))
```

```python
mais_similares('snape',harry_wv)
mais_similares('dumbledore',harry_wv)
```

## Intruso e Operações

```python
def intruso(lista,modelo):
    return(modelo.wv.doesnt_match(lista))
```

```python
def analogy(x2, y1, x1,modelo):
    result = modelo.wv.most_similar(positive = [y1,x2], negative = [x1])
    return result[0][0]
```

```python
lista = ['dumbledore','harry','prenda']
print(intruso(lista,harry_wv))
```
